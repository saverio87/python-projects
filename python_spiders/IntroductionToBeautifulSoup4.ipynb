## Extract the job details wrapper

```python
#span result-hood
#a result-title hdrlnk
#time result-date :: text
```

```python
jobs = soup.find_all("p",{"class":"result-info"})
```

```python
for job in jobs:
    title = job.find('a', {"class":"result-title hdrlnk"}).text
    location_tag = job.find('span',{"class":"result-hood"})
    location = location_tag.text if location_tag else "N A"
    time = job.find('time',{"class":"result-date"}).text
    link = job.find('a', {"class":"result-title hdrlnk"}).get('href')
    print("Job title: " + title + "\nLocation: " + location + "\nTime: " + time + "\nLink: " + link + "\n---")
```

```python
## Here's a cool way to create a list with all the URLs you are going to need to scrape
pages = []
pages_to_scrape=41
for i in range(20,pages_to_scrape):
    url = ('http://www.esl-teachersboard.com/cgi-bin/intl/index.pl?page={}').format(i);
    pages.append(url);
```

```python
for item in pages:
    page = requests.get(item);
    soup = BeautifulSoup(page.text, 'html.parser');
    
    for tag in soup.select("dd.thread a", href=True):
        url_to_scrape = ('http://www.esl-teachersboard.com{}').format(tag.get('href'));
        allLinks.append(url_to_scrape);      
```

```python
allLinks[0]
```

```python
for link in allLinks:
    page = requests.get(link);
    soup = BeautifulSoup(page.text, 'html.parser');
    
    main_div = soup.find_all("div",{"id":"msg_wrap"})
    
    for i in main_div:
        jobLocation = main_div.find('div',{'class':'msg_headln'}).text;
        location.append(jobLocation);
        adPoster = main_div.find('span',{'class':'msg_poster'}).text;
        school_or_agency.append(adPoster);
    ## emailAddress = main_div.find('span', {"class":"msg_email"});
    ## email.append(emailAddress.findChildren("a")) if emailAddress else email.append('N A');
        adDate = main_div.find('span',{'class':'msg_date'}).text;
        date.append(adDate);
        jobAd = main_div.find('div',{'class':'msg_text'}).text;
        job_ad.append(jobAd);
```
